```python
from utils import misc as misc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



ticker_list = ['UPRO', 'TECL', 'FAS', 'TMF', 'USO', 'DBA', 'GLD', 'DBB', 'EEM', 'BITO']

for ticker in ticker_list:
    print(ticker)
    df = misc.get_price_data(ticker)
    df.to_csv('datasets/stock_price_series/' + ticker + '.csv')

```

```python
port_val = 5000

def update(df):
    # df = misc.clean_df(df)
    df['20_up_bound'] = df['high'].rolling(20).max()
    df['20_low_bound'] = df['low'].rolling(20).min()
    df['10_up_bound'] = df['high'].rolling(10).max()
    df['10_low_bound'] = df['low'].rolling(10).min()

    df['High-Low'] = df['high'] - df['low']
    df['High-PrevClose'] = abs(df['high'] - df['close'].shift(1))
    df['Low-PrevClose'] = abs(df['low'] - df['close'].shift(1))
    df['TR'] = df[['High-Low', 'High-PrevClose', 'Low-PrevClose']].max(axis=1)
    df['ATR'] = df['TR'].rolling(20).mean()
    df = df.dropna()

for ticker in ticker_list:
    print(ticker)
    df = pd.read_csv('datasets/stock_price_series/'+ ticker + '.csv')
    df = misc.clean_df(df)
    update(df)
    print(f'date:           {str(df["date"].to_list()[-1])[:11]}')
    print(f'last_price:     {np.round(df["close"].to_list()[-1], 2)}')
    n = np.round(df["ATR"].to_list()[-1], 2)
    print(f'3 x ATR:        {3 * n}')
    print(f'risk unit qty:  {int(port_val * 0.01 / (3 * n))}')
    print('---')
    print(f'long entry:     {np.round(df["20_up_bound"].to_list()[-1], 2)}')
    print(f'long exit:      {np.round(df["10_low_bound"].to_list()[-1], 2)}')
    print('---')
    print(f'short entry:    {np.round(df["20_low_bound"].to_list()[-1], 2)}')
    print(f'short exit:     {np.round(df["10_up_bound"].to_list()[-1], 2)}')
    print('----------------------------------------')
    
```
